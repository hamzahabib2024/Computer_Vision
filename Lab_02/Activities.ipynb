```python
import cv2
print(cv2.__version__)
```

```python
image = cv2.imread("E:\\vs code\\git\\Computer Vision\\cameraman.png")

image_copy = image.copy()

img_circle = cv2.circle(image_copy, (150,100), 80, (0,255,0), 4)
cv2.imshow("cameraman",image)
cv2.imshow("cameraman",img_circle)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Drawing lines

```python
image = cv2.imread("E:\\vs code\\git\\Computer Vision\\cameraman.png")
image_copy = image.copy()

print(image_copy.shape)
img_line= cv2.line(image_copy, (150,100), (300,800), (0,255,0),4)

cv2.imshow("img_line",img_line)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
