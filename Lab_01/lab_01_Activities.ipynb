# Activity 02
    Perform Basic Image Operations :
        1. Resize the Image 
        2. Rotate the Image 
        3. Flip the Image Horizontally

```python
# for cameraman picture

image = cv2.imread("E:\\vs code\\git\\Computer Vision\\Lab_01\\cameraman.png") 

# cv2.resize() ---> is used to resize an image to specified width and height
resized_image = cv2.resize(image, (400, 300))

# cv2.rotate() ---> is used to rotate an image by specified angle
#there are 4 types of rotation: 90 degree clockwise, 90 degree counterclockwise, 180 degree 
# types are: cv2.ROTATE_90_CLOCKWISE, cv2.ROTATE_90_COUNTERCLOCKWISE, cv2.ROTATE_180
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

# cv2.flip() ---> is used to flip an image either horizontally, vertically, or both
# flipCode: 0 = vertical, 1 = horizontal, -1 = both
# here we are flipping the image horizontally
flipped_image = cv2.flip(image, 1)
# flipped_image = cv2.flip(image, 0)
# flipped_image = cv2.flip(image, -1)

cv2.imshow("Resized Image", resized_image)
cv2.imshow("Rotated Image", rotated_image)
cv2.imshow("Flipped Image", flipped_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# for lena picture

image  = cv2.imread("E:\\vs code\\git\\Computer Vision\\Lab_01\\lena.jpg")

resized_image = cv2.resize(image, (400, 300))
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
flipped_image = cv2.flip(image,1)

cv2.imshow("Original Image", image)

cv2.imshow("Resized Image", resized_image)
cv2.imshow("Rotated Image", rotated_image)
cv2.imshow("Flipped Image", flipped_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Activity 03
    Apply Image Filtering Techniques 
        1. Apply Gaussian Blur
        2. Convert Image to Grayscale

```python
# for cameraman picture

image = cv2.imread("E:\\vs code\\git\\Computer Vision\\Lab_01\\cameraman.png")

# gaussianBlur() ---> is used to apply Gaussian Blur to an image
# it takes three parameters: source image, kernel size, and standard deviation in X direction
# kernel size is represented as (width, height) and both width and height should be positive and odd numbers
# 0 is used to automatically calculate the standard deviation based on the kernel size
blurred_image = cv2.GaussianBlur(image, (7, 7), 0)

# cvtColor() ---> is used to convert an image from one color space to another
# here we are converting the image from BGR to Grayscale
# cvtColor() takes two parameters: source image and color conversion code
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow("Blurred Image", blurred_image)
cv2.imshow("Grayscale Image", gray_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# for lena picture

image = cv2.imread("E:\\vs code\\git\\Computer Vision\\Lab_01\\lena.jpg")

blurred_image = cv2.GaussianBlur(image, (25,25), 0)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow("Original Image", image)
cv2.imshow("Blurred Image", blurred_image)
cv2.imshow("GreyScale Image", gray_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Activity 04
    Draw on Images to Annotate and Highlight Features 
        1. Draw a Rectangle 
        2. Add Text

```python
# for cameraman picture

image = cv2.imread("E:\\vs code\\git\\Computer Vision\\Lab_01\\cameraman.png")

# image.copy() ---> is used to create a copy of an image
# this is useful when we want to annotate an image without modifying the original image
# annotate means to add shapes, text, or other information to an image
annotated_image = image.copy()

# cv2.rectangle() ---> is used to draw a rectangle on an image
# it takes five parameters: image, top-left corner, bottom-right corner, color, and thickness
# color is represented as (B, G, R)
# 2nd parameter is the position of the top-left corner of the rectangle (x1, y1) and here (50, 50). it is the starting point of rectangle
# 3rd parameter is the position of the bottom-right corner of the rectangle (x2, y2) and here (300, 200). it is the ending point of rectangle
# thickness is an integer value representing the thickness of the rectangle border
cv2.rectangle(annotated_image, (50, 50), (300, 200), (0, 255, 0), 2)


# cv2.putText() ---> is used to put text on an image
# it takes several parameters: image, text, position, font, font scale, color, and thickness
# the 3rd parameter is the position (x, y) where the text will start. here (60, 45)
# font scale is a float value representing the size of the text
cv2.putText(
    annotated_image,
    "Hello, OpenCV!",
    (60, 45),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (255, 0, 0),
    2
)

cv2.imshow("Annotated Image", annotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# for lena picture

image = cv2.imread("E:\\vs code\\git\\Computer Vision\\Lab_01\\lena.jpg")

annotated_image = image.copy()

# cv2.circle() ---> is used to draw a circle on an image
# it takes five parameters: image, center coordinates, radius, color, and thickness
# center coordinates means the position (x, y) of the center of the circle
# the radius is in scale of pixels.
cv2.circle(annotated_image, (260,215),  200, (0,0,255), 6)

cv2.putText(
    annotated_image,
    "Circle around Lena",
    (150, 30),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (255, 255, 255),
    2
)

cv2.imshow("Circle on Image", annotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
